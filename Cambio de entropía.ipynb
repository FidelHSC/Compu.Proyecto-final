$\Delta{S} = nRln\left(\frac{V_f}{V_0}\right)$ 

$R = 8.314472\ \frac{J}{molK}$ 

```python
#Importamos bibliotecas para graficar, manejar listas y funciones matemáticas, respectivamnete
import matplotlib.pyplot as plt
import numpy as np
import math 
```

```python
#Variables que el usuario puede definir
n = float(input("Ingrese el número de moles de su gas ideal: "))
V = float(input("Ingrese un volúmen (m^3) total: "))
```

```python
#Creamos listas vacías que se llenarán gracias a un ciclo for y la función appned de numpy
LDeltaS =[]
LV = []
R = 8.314472

for v in np.arange(0.1,V+1):
    DeltaS = n*R*(math.log1p(v))
    
    LDeltaS.append(DeltaS)
    LV.append(v)    
```

```python
#Graficamos la función deseada para el cambio de entropía
plt.plot(LV, LDeltaS, label = "Cambio de entropía")
plt.title("Cambio de entropía vs volúmen")
plt.xlabel("Volúmen (V)")
plt.ylabel("Cambio de entropía (DeltaS)")
plt.legend(loc = 2)
```

```python
#Usamos las variables anteriores, únicamnete definir a T y procedemos a graficar la isoterma, el calor y el trabajo
T = float(input("Ingrese un valor para la tempreratura (K) constante: "))
```

```python
#Hacemos el mismo procedimiento que con el ciclo for previo
LP = []

for v in np.arange(0.1,V+1):
    P = (n*R*T)/v
    
    LP.append(P)
```

```python
#Graficamos
plt.plot(LV, LP,'r', label = "Isoterma")
plt.title("Diagrama P-V")
plt.xlabel("Volúmen (V)")
plt.ylabel("Presión (P)")
plt.legend(loc = 1)
```

```python
#Una vez más utilizamos este método para graficar funciones
LW_Q =[]
for v in np.arange(0.1,V+1):
    W_Q = n*R*(math.log1p(v))
    
    LW_Q.append(W_Q)
```

```python
#Graficamos de nuevo
plt.plot(LV, LW_Q, 'y', label = "Trabajo/Calor")
plt.title("Trabajo/Calor vs volúmen")
plt.xlabel("Volúmen (V)")
plt.ylabel("Trabajo/Calor (J)")
plt.legend(loc = 2)
```
